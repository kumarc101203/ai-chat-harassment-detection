```python
import pandas as pd

data = pd.read_csv(r"D:\hackathons & other projects\ai-chat-harassment-detection\data\raw\train.csv")

print("Shape:", data.shape)
data.head()
```

```python
# create single harassment label

toxicity_columns = [
    "toxic",
    "severe_toxic",
    "obscene",
    "threat",
    "insult",
    "identity_hate"
]

data["harassment"] = data[toxicity_columns].max(axis=1)

# keep only required columns

data = data[["comment_text", "harassment"]]

print(data.head())

print("\nDistribution:")
print(data["harassment"].value_counts())
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    data,
    test_size=0.1,
    stratify=data["harassment"],
    random_state=42
)

print("Train shape:", train_df.shape)
print("Validation shape:", val_df.shape)

print("\nTrain distribution:")
print(train_df["harassment"].value_counts(normalize=True))

print("\nValidation distribution:")
print(val_df["harassment"].value_counts(normalize=True))
```

```python
train_df.to_csv("../data/processed/train_processed.csv", index=False)
val_df.to_csv("../data/processed/val_processed.csv", index=False)
```
